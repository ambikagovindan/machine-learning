```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
def get_score(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
get_score(LogisticRegression(),X_train,X_test,y_train,y_test)
```

```python
get_score(DecisionTreeClassifier(),X_train,X_test,y_train,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score_lr = []
score_svc = []
score_tree = []
```

```python
score_lr.append(cross_val_score(LogisticRegression(),digits.data,digits.target))
score_svc.append(cross_val_score(SVC(),digits.data,digits.target))
score_tree.append(cross_val_score(DecisionTreeClassifier(),digits.data,digits.target))
```

```python
score_lr
```

```python
mean_score_lr = np.mean(score_lr)
mean_score_svc = np.mean(score_svc)
mean_score_tree= np.mean(score_tree)
mean_score_lr
```
